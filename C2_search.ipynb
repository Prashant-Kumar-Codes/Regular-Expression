```python
import re
```

- **`re.search(pattern, string)`**  
  Looks through the entire string to find the first place where the given pattern appears.  
  -> In your code, it searches for the word `"Prashant"` inside the text.
  
  NOTE - RETURNS NONE IF NOTHING MATCHED

- **`match.group()`**  
  Returns the exact part of the string that matched the pattern.  
  -> Here, it will print `"Prashant"` because that’s the word found.

- **`match.start()`**  
  Gives the starting index (position) of the matched substring in the original text.  
  -> For `"Prashant"`, it shows where the word begins.

- **`match.end()`**  
  Gives the ending index (position) of the matched substring in the original text (just after the last character).  
  -> For `"Prashant"`, it shows where the word finishes.


```python
txt = 'Hello this is Prashant Kumar. I am Python programmer and AI/ML engineer.'
```

```python
for i in range(len(txt)):
    if txt[i] == '.':
        print(i)
```

```python
match = re.search(r'Prashant', txt)
print(match)
print(match.group())
print(match.start())
print(match.end())
```

\ :	Used to drop the special meaning of character following it

\\ → Escapes special meaning.
Example: re.search(r"\.", "a.b") → matches the literal ..

```python
match = re.search('t', txt)
print(match.group())
```

```python
match = re.search(r'.', txt)
print(match)
print(match.group())

match = re.search(r'\.', txt)
print(match)
print(match.group())
```
